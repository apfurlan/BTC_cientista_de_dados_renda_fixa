1. Funções para busca e tratamento de dados do tesouro direto

```python
def busca_titulos_tesouro_direto():
    url = 'https://www.tesourotransparente.gov.br/ckan/dataset/df56aa42-484a-4a59-8184-7676580c81e3/resource/796d2059-14e9-44e3-80c9-2d9e30b405c1/download/PrecoTaxaTesouroDireto.csv'
    df = pd.read_csv(url, sep=';', decimal=',')
    df['Data Vencimento'] = pd.to_datetime(df['Data Vencimento'], dayfirst=True)
    df['Data Base'] = pd.to_datetime(df['Data Base'], dayfirst=True)
    multi_indice = pd.MultiIndex.from_frame(df.iloc[:,:3])
    df = df.set_index(multi_indice).iloc[:, 3:]
    return df
```

1. Buscando dados

```python
titulos = busca_titulos_tesouro_direto()
titulos
```

2. Salvando CSV

```python
data_venc = '2025-01-01'
prefixado_2025 = titulos.loc[('Tesouro Prefixado', data_venc)]
```

```python
prefixado_2025.to_csv('prefixado_2025.csv')
```

3. Salvando diversos arquivos

```python
lista_titulos = {'Tesouro Prefixado': ['2025-01-01', '2029-01-01'], 'Tesouro IPCA+': ['2026-08-15', '2035-05-15', '2045-05-15']}
```

```python
for titulo, vencimentos in lista_titulos.items():
    for vencimento in vencimentos:
        nome_titulo = titulo.replace(' ', '_').lower()
        ano_titulo = str(vencimento)[0:4]
        nome_arquivo = nome_titulo + '_' + ano_titulo + '.csv'
        dados_titulos = titulos.loc[(titulo, vencimento)]
        dados_titulos.to_csv(nome_arquivo)
```
